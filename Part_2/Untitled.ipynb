1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.

    • Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.
    
    • Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.
    
    • Исключим стоп-слова с помощью stop_words='english'.
    
    • Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью 
    CountVectorizer.get_feature_names().

```python
# Преобразуем данные
df['tweet_stemmed'] = df['tweet_stemmed'].apply(lambda tokens: ' '.join(tokens))
df['tweet_lemmatized'] = df['tweet_lemmatized'].apply(lambda tokens: ' '.join(tokens))
df.head()
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 1), # quantity colocations
                                   analyzer='word', # 'char'
                                   binary=False, # or all quantity
                                   tokenizer=str.split, 
                                   stop_words="english", 
                                   max_df=0.9, 
                                   max_features=1000)
count_vectorizer
```

```python
#Создаем the Bag-of-Words модель для tweet_stemmed
bag_of_words_stemmed = count_vectorizer.fit_transform(df['tweet_stemmed'])
bag_of_words_stemmed
```

```python
# Отобразим Bag-of-Words модель как DataFrame
columns = count_vectorizer.get_feature_names()
stemmed_count = pd.DataFrame(bag_of_words_stemmed.toarray(), columns=columns)
stemmed_count.head()
```

```python
# Создаем the Bag-of-Words модель для tweet_lemmatized
bag_of_words_lemmatized = count_vectorizer.fit_transform(df['tweet_lemmatized'])
bag_of_words_lemmatized
```

```python
# Отобразим Bag-of-Words модель как DataFrame
columns = count_vectorizer.get_feature_names()
lemmatized_count = pd.DataFrame(bag_of_words_lemmatized.toarray(), columns=columns)
lemmatized_count.head()
```
