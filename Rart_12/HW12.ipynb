```python
from datasets import load_dataset

corpus = load_dataset('merionum/ru_paraphraser')
```

```python
corpus
```

```python
import pandas as pd

df_train = pd.DataFrame(corpus['train'])
df_train
```

```python
df_train['class'].value_counts()
```

```python
df_test = pd.DataFrame(corpus['test'])
df_test
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
import tensorflow as tf
```

Возьмем мультилингвальную бертовую модель для классификации на 3 класса:

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=3)
```

```python
model.summary()
```

Токенизируем и создадим тф-датасет:

```python
def tokenize_function(examples):
    return tokenizer(examples["text_1"], examples["text_2"], padding="max_length", truncation=True, return_tensors='tf')
```

```python
tokenized_datasets = corpus.map(tokenize_function, batched=True)
```

```python
full_train_dataset = tokenized_datasets["train"]
full_eval_dataset = tokenized_datasets["test"]
```

```python
full_train_dataset
```

```python
tf_train_dataset = full_train_dataset.remove_columns(['id', 'id_1', 'id_2', 'text_1', 'text_2']).with_format("tensorflow")
tf_eval_dataset = full_eval_dataset.remove_columns(['id', 'id_1', 'id_2', 'text_1', 'text_2']).with_format("tensorflow")
```

```python
tf_train_dataset['input_ids']
```

```python
num_classes = 3
y_train = tf.keras.utils.to_categorical(tf_train_dataset["class"], num_classes)
y_eval = tf.keras.utils.to_categorical(tf_eval_dataset["class"], num_classes)
```

```python
y_train
```

```python
train_features = {x: tf_train_dataset[x] for x in tokenizer.model_input_names}
train_tf_dataset = tf.data.Dataset.from_tensor_slices((train_features, y_train))
train_tf_dataset = train_tf_dataset.shuffle(len(tf_train_dataset)).batch(4)

eval_features = {x: tf_eval_dataset[x] for x in tokenizer.model_input_names}
eval_tf_dataset = tf.data.Dataset.from_tensor_slices((eval_features, y_eval))
eval_tf_dataset = eval_tf_dataset.batch(4)
```

```python
model.layers
```

Заморозим для обучения бертовый слой и обучим только выходной плотный слой:

```python
model.layers[0].trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
    metrics=[tf.metrics.CategoricalAccuracy('accuracy')],
)

model.summary()
```

```python
model.fit(train_tf_dataset, validation_data=eval_tf_dataset, epochs=3)
```

```python
from tensorflow.keras.callbacks import Callback
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, classification_report
import numpy as np


class Metrics(Callback):


    def on_test_end(self, epoch, logs=None):
        logits_to_class = {0:0, 1:1, 2:-1}
        val_predict = model.predict(eval_tf_dataset).logits.argmax(axis=1)
        val_predict_lb = np.array([logits_to_class[i] for i in val_predict])
        print('\n', classification_report(df_test['class'].astype('int32'), val_predict_lb))
```

Посмотрим отчет классификации на тестовом датасете:

```python
model.evaluate(eval_tf_dataset, callbacks=[Metrics()])
```

**Вывод:**

Датасет имеет 3 класса: -1 - предложения не идентичны, 1 - предложения полностью идентичны и 0 - предложения почти идентичны. Как видно, модель крайне плохо улавливает такую постановку задачи. Путает между собой полную и неполную идентичность, а также практически не улавливает неидентичность предложений.

Попробуем оставить 2 класса - предложения идентичны или неидентичны. Классы 0 и 1 сольем в один класс:

```python
convert_classes = {'-1': 0, '0': 1, '1': 1}
y_train2 = df_train['class'].map(convert_classes)
y_eval2 = df_test['class'].map(convert_classes)
```

```python
num_classes = 2
y_train2 = tf.keras.utils.to_categorical(y_train2, num_classes)
y_eval2 = tf.keras.utils.to_categorical(y_eval2, num_classes)
```

```python
y_train2
```

Возьмем такую же модель для классификации, но с 2-мя классами:

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=2)
```

```python
train_features = {x: tf_train_dataset[x] for x in tokenizer.model_input_names}
train_tf_dataset = tf.data.Dataset.from_tensor_slices((train_features, y_train2))
train_tf_dataset = train_tf_dataset.shuffle(len(tf_train_dataset)).batch(64)

eval_features = {x: tf_eval_dataset[x] for x in tokenizer.model_input_names}
eval_tf_dataset = tf.data.Dataset.from_tensor_slices((eval_features, y_eval2))
eval_tf_dataset = eval_tf_dataset.batch(64)
```

```python
model.layers[0].trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
    metrics=[tf.metrics.CategoricalAccuracy('accuracy')],
)

model.summary()
```

Также заморозили берт, обучим модель:

```python
model.fit(train_tf_dataset, validation_data=eval_tf_dataset, epochs=3)
```
